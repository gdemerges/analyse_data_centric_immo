```python
import matplotlib.pyplot as plt

# Calculer l'explosion pour chaque tranche
explode = (0, 0, 0, 0,0,0)  # Décalage seulement pour la troisième tranche (indice 2)

plt.pie(nature_mutation,autopct='%1.1f%%', explode = explode)

# Ajout d'un titre
plt.title('Répartition des types de mutation')
plt.axis('equal')
# Affichage du pie chart
plt.show()
```
